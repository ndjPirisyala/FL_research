```python
import pandas as pd
```

```python
df1 = pd.read_csv('Test1.csv')
df2 = pd.read_csv('Test3.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.plot(df1['No. of clients'],df1['IID 100'],label='without weights, batch size 100')
plt.plot(df2['No. of clients'],df2['IID 100'],label='with weights(std_deviation), batch size 100')
plt.plot(df1['No. of clients'],df1['IID 1000'],label='without weights, batch size 1000')
plt.plot(df2['No. of clients'],df2['IID 1000'],label='with weights(std_deviation), batch size 1000')
plt.xlabel('Number of clients', fontsize=10)
plt.ylabel('Accuracy', fontsize=10)
plt.legend()
plt.savefig("IID.png", dpi=100)
plt.show()
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.plot(df1['No. of clients'],df1['non-IID 100'],label='without weights, batch size 100')
plt.plot(df2['No. of clients'],df2['non-IID 100'],label='with weight(std), batch size 100')
plt.plot(df1['No. of clients'],df1['non-IID 1000'],label='without weights, batch size 1000')
plt.plot(df2['No. of clients'],df2['non-IID 1000'],label='with weight(std), batch size 1000')
plt.xlabel('Number of clients', fontsize=10)
plt.ylabel('Accuracy', fontsize=10)
plt.legend()
plt.savefig("NON_IID_without_weight_with_weight_std.png", dpi=100)
plt.show()
```
