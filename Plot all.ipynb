```python
import seaborn as sns
from matplotlib import pyplot as plt
from pickle import load
sns.set_theme(style= 'whitegrid')
```

```python
accuracy_list_without_wights = load(open('accuracy_list_without_wights','rb'))
accuracy_list_with_wights1 = load(open('accuracy_list_with_wights1','rb'))    
accuracy_list_with_wights2 = load(open('accuracy_list_with_wights2','rb'))    
accuracy_list_with_wights3 = load(open('accuracy_list_with_wights3','rb'))    
accuracy_list_with_wights4 = load(open('accuracy_list_with_wights4','rb'))    
accuracy_list_with_wights5 = load(open('accuracy_list_with_wights5','rb'))    
accuracy_list_with_wights6 = load(open('accuracy_list_with_wights6','rb'))    
```

```python
# plt.title('Poisson distribution ')
fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(range(1,len(accuracy_list_without_wights)),accuracy_list_without_wights[1:],label="No Weighted")
plt.plot(range(1,len(accuracy_list_with_wights1)),accuracy_list_with_wights1[1:],label="Min mean")
plt.plot(range(1,len(accuracy_list_with_wights2)),accuracy_list_with_wights2[1:],label="Max mean")
plt.plot(range(1,len(accuracy_list_with_wights3)),accuracy_list_with_wights3[1:],label="sum of mean")
plt.plot(range(1,len(accuracy_list_with_wights4)),accuracy_list_with_wights4[1:],label="sum of squared mean")
plt.plot(range(1,len(accuracy_list_with_wights5)),accuracy_list_with_wights5[1:],label="root sum of squared mean")
plt.plot(range(1,len(accuracy_list_with_wights6)),accuracy_list_with_wights6[1:],label="random")

plt.legend(loc="lower right")
plt.ylabel('loss')
plt.xlabel('rounds')
plt.show()
```

```python
from scipy import interpolate

RANGE = range(len(accuracy_list_without_wights[1:]))

fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_without_wights[1:])(RANGE), label="No Weighted")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights1[1:])(RANGE),label="Max mean")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights2[1:])(RANGE),label="Min mean")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights3[1:])(RANGE),label="sum of mean")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights4[1:])(RANGE),label="sum of squared mean")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights5[1:])(RANGE),label="root sum of squared mean")
plt.plot(range(1,len(accuracy_list_without_wights)),interpolate.UnivariateSpline(RANGE, accuracy_list_with_wights6[1:])(RANGE),label="random")

plt.legend(loc="lower right")
plt.ylabel('loss')
plt.xlabel('rounds')
plt.show()
```
