```python
import glob
import pandas as pd
from collections import defaultdict as ddict
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set_style("whitegrid")
```

```python
files = glob.glob('../hand_written_digits/*')
```

```python
files
```

```python
data = [list(map(float, i.split())) for i in open(files[4],'r').readlines()]
```

```python
len(data[0])
```

```python
DF = ddict(list)
for i in data:
    DF['avg'].append(sum(i)/240)
DF = pd.DataFrame(DF)
```

```python
DF
```
