```python
def CalMLE(m,b,p,q):
    return pow(q-(m*p+b),2)
```

```python
import math
def MLE(m,b,X,Y):
    return math.sqrt(sum(list(map(lambda x:CalMLE(m,b,x[0],x[1]),zip(X,Y)))))
```

```python
# y=1x+2
m1 = 1.3
b1 = 1
CL1 = lambda x:m1*x+b1
# y= -x
m2=1
b2=-2
CL2 = lambda x:m2*x+b2
```

```python
#error
MLE(m1,b1,x1,y1)
```

```python
#error
MLE(m2,b2,x2,y2)
```
