```python
from matplotlib import pyplot as plt
import pickle
from scipy import interpolate


import seaborn as sns
sns.set_theme(style= 'whitegrid')

agg1 = pickle.load(open('agg1','rb'))
agg2 = pickle.load(open('agg2','rb'))

RANGE = range(len(agg1))

L1 = list(map(lambda x:x[0],agg1))
L2 = list(map(lambda x:x[0],agg2))

A1 = list(map(lambda x:x[1],agg1))
A2 = list(map(lambda x:x[1],agg2))
```

```python
fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(L1, label="No Weighted")
plt.plot(L2, label="Weighted")
plt.legend(loc="lower right")
plt.ylabel('loss')
plt.xlabel('rounds')
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(range(len(L1)),interpolate.UnivariateSpline(RANGE, L1)(RANGE), label="No Weighted")
plt.plot(range(len(L2)),interpolate.UnivariateSpline(RANGE, L2)(RANGE), label="Weighted")
plt.legend(loc="lower right")
plt.ylabel('loss')
plt.xlabel('rounds')
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(A1, label="NoWeighted")
plt.plot(A2,  label="Weighted")
plt.legend(loc="lower right")
plt.ylabel('accuracy')
plt.xlabel('rounds')
plt.show()
```

```python
fig = plt.figure(num=None, figsize=(8, 6))
plt.plot(range(len(A1)),interpolate.UnivariateSpline(RANGE, A1)(RANGE), label="No Weighted")
plt.plot(range(len(A2)),interpolate.UnivariateSpline(RANGE, A2)(RANGE), label="Weighted")
plt.legend(loc="lower right")
plt.ylabel('accuracy')
plt.xlabel('rounds')
plt.show()
```
