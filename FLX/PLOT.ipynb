```python
from matplotlib import pyplot as plt
import pickle
from scipy import interpolate


import seaborn as sns
sns.set_theme(style= 'whitegrid')

RANGE = range(20)

agg1 = pickle.load(open('agg1','rb'))
# agg = cmap(agg)
L1 = list(map(lambda x:x[0],agg1))
A1 = list(map(lambda x:x[1],agg1))

# # 2
agg2 = pickle.load(open('agg2','rb'))
# agg2 = cmap(agg2)
L2 = list(map(lambda x:x[0],agg2))
A2 = list(map(lambda x:x[1],agg2))

# # 3
agg3 = pickle.load(open('agg3','rb'))
# agg3 = cmap(agg3)
L3 = list(map(lambda x:x[0],agg3))
A3 = list(map(lambda x:x[1],agg3))

# # 4
agg4 = pickle.load(open('agg4','rb'))
# agg4 = cmap(agg4)
L4 = list(map(lambda x:x[0],agg4))
A4 = list(map(lambda x:x[1],agg4))

# # 5
# agg5 = pickle.load(open('agg5','rb'))
# agg5 = cmap(agg5)
# L5 = list(map(lambda x:x[0],agg5))
# A5 = list(map(lambda x:x[1],agg5))

# # # 6
# agg6 = pickle.load(open('agg6','rb'))
# agg6 = cmap(agg6)
# L6 = list(map(lambda x:x[0],agg6))
# A6 = list(map(lambda x:x[1],agg6))
```

```python
fig, axs = plt.subplots(1, 2)
fig.tight_layout()

axs[0].plot(range(len(L1)),interpolate.UnivariateSpline(RANGE, L1)(RANGE), label="No Weighted")
axs[0].plot(range(len(L2)),interpolate.UnivariateSpline(RANGE, L2)(RANGE), label="Weighted")
axs[0].set_ylabel('loss')
axs[0].set_xlabel('rounds')

axs[1].plot(range(len(A1)),interpolate.UnivariateSpline(RANGE, A1)(RANGE), label="No Weighted")
axs[1].plot(range(len(A2)),interpolate.UnivariateSpline(RANGE, A2)(RANGE), label="Weighted")
axs[1].set_ylabel('accuracy')
axs[1].set_xlabel('rounds')

plt.legend(loc="lower right")
plt.show()
```
